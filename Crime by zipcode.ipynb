```python
import pandas as pd
import numpy as np
import csv
import geopy
```

```python
df = pd.read_csv('sampled crime.csv')
```

```python
def zipcode(df, geolocator, lat_field, lon_field):
    location = geolocator.reverse((df[lat_field], df[lon_field]))
    address = location.raw['address']
    if 'postcode' in address.keys():
        return location.raw['address']['postcode']
    else:
        return None

geolocator = geopy.Nominatim(user_agent='JoanneX')
```

```python
newdf = df[['Latitude','Longitude']]
newdf.head()
```
