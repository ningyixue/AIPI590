```python
import pandas as pd
import numpy as np
import csv
```

```python
df = pd.read_csv('NYPD_Complaint_Data_Historic-2.csv')
df.head()
```

```python
# Choose specific columns
df_new = df[['CMPLNT_FR_DT','Latitude','Longitude','PATROL_BORO']]
df_new.head()
```

```python
#drop null values
df_new = df_new.dropna()
```

```python
#only keep 2019 data
df_new['CMPLNT_FR_DT'] = df_new['CMPLNT_FR_DT'].str.split('/').str.get(2)
df_final = df_new[df_new['CMPLNT_FR_DT'] == '2019']
```

```python
#sample only 10000 data
df_sample= df_final.sample(n=10000)
df_sample
```

```python
#export to csv
Result ='/Users/tylin/Downloads/sampled crime.csv'

df_sample.to_csv( Result, index=False )
```
